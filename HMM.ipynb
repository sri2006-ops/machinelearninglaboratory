```python
# Install library
!pip install hmmlearn ipywidgets --quiet

# -------------------------------
# Imports
# -------------------------------
import numpy as np
from hmmlearn.hmm import CategoricalHMM
import ipywidgets as widgets
from IPython.display import display

# -------------------------------
# Step 1: Create HMM Model
# -------------------------------
# 2 Hidden States: Healthy (0), Sick (1)
model = CategoricalHMM(n_components=2)

# Start Probabilities
model.startprob_ = np.array([0.6, 0.4])

# Transition Probabilities
model.transmat_ = np.array([
    [0.7, 0.3],   # Healthy → Healthy/Sick
    [0.4, 0.6]    # Sick → Healthy/Sick
])

# Emission Probabilities
# Observations: 0=Normal, 1=Cold, 2=Dizzy
model.emissionprob_ = np.array([
    [0.5, 0.4, 0.1],   # Healthy
    [0.1, 0.3, 0.6]    # Sick
])

# -------------------------------
# Step 2: UI Components
# -------------------------------
obs_map = {"Normal": 0, "Cold": 1, "Dizzy": 2}

obs1 = widgets.Dropdown(options=list(obs_map.keys()), description="Obs1:")
obs2 = widgets.Dropdown(options=list(obs_map.keys()), description="Obs2:")
obs3 = widgets.Dropdown(options=list(obs_map.keys()), description="Obs3:")

button = widgets.Button(description="Predict Hidden States", button_style='success')
output = widgets.Output()

# -------------------------------
# Step 3: Prediction Function
# -------------------------------
def predict(b):
    with output:
        output.clear_output()

        # Convert input to numbers
        sequence = [obs_map[obs1.value], obs_map[obs2.value], obs_map[obs3.value]]
        sequence = np.array(sequence).reshape(-1, 1)

        # Predict hidden states using Viterbi
        logprob, states = model.decode(sequence)

        state_names = ["Healthy", "Sick"]

        print("🔍 Input Observations:", obs1.value, obs2.value, obs3.value)
        print("\n🧠 Predicted Hidden States:")

        for i, s in enumerate(states):
            print(f"Step {i+1}: {state_names[s]}")

        print(f"\n📊 Log Probability: {round(logprob, 2)}")

# Button click
button.on_click(predict)

# -------------------------------
# Step 4: Display UI
# -------------------------------
display(obs1, obs2, obs3, button, output)
```
