```python
# Install required libraries
!pip install scikit-learn ipywidgets matplotlib --quiet

# -------------------------------
# Imports
# -------------------------------
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import ipywidgets as widgets
from IPython.display import display

# -------------------------------
# Step 1: Generate Dataset
# -------------------------------
def generate_data(n_samples, centers):
    X, _ = make_blobs(n_samples=n_samples, centers=centers, random_state=42)
    return X

# -------------------------------
# Step 2: UI Components
# -------------------------------
samples_slider = widgets.IntSlider(value=200, min=50, max=500, step=50, description="Samples")
clusters_slider = widgets.IntSlider(value=3, min=2, max=6, step=1, description="Clusters")
init_dropdown = widgets.Dropdown(options=["k-means++", "random"], value="k-means++", description="Init")
run_button = widgets.Button(description="Run Clustering", button_style='success')

output = widgets.Output()

# -------------------------------
# Step 3: Clustering Function
# -------------------------------
def run_clustering(b):
    with output:
        output.clear_output()

        n_samples = samples_slider.value
        n_clusters = clusters_slider.value
        init_method = init_dropdown.value

        # Generate data
        X = generate_data(n_samples, n_clusters)

        # Apply KMeans
        model = KMeans(n_clusters=n_clusters, init=init_method, random_state=42)
        labels = model.fit_predict(X)
        centers = model.cluster_centers_

        # -------------------------------
        # Visualization
        # -------------------------------
        plt.figure()
        plt.scatter(X[:, 0], X[:, 1], c=labels)
        plt.scatter(centers[:, 0], centers[:, 1], marker='X', s=200)
        plt.title("K-Means Clustering Result")
        plt.xlabel("Feature 1")
        plt.ylabel("Feature 2")
        plt.show()

        # -------------------------------
        # Output Info
        # -------------------------------
        print("📊 Clustering Completed!")
        print(f"Number of Samples: {n_samples}")
        print(f"Number of Clusters: {n_clusters}")
        print(f"Initialization: {init_method}")

# Button click
run_button.on_click(run_clustering)

# -------------------------------
# Step 4: Display UI
# -------------------------------
display(samples_slider, clusters_slider, init_dropdown, run_button, output)
```
